## Features

- `접수월(RECEIPT_MT)`: 비씨카드가 카드사용내역을 접수한 월


- `가맹점업종분류명_대(MRHST_INDUTY_CL_NM_LRGE)`: 비씨카드 가맹점 업종 대분류명(총 42개업종 대분류)


- `가맹점업종분류명(MRHST_INDUTY_CL_NM)`: 비씨카드 가맹점 업종 중분류명(총 350개 업종 중분류)


- `연령대코드(AGRDE_CODE)`: 10세기준 연령대 코드


- `성별코드(SEXDSTN_CODE)`: 0: 미기재, 1: 남성, 2: 여성, 9: 기타


- `광역시도명(BRTC_NM)`: 가맹점 우편번호 기준 광역시도명


- `시군구명(SIGNGU_NM)`: 가맹점 우편번호 기준 시군구명


- `금액지표(AMOUNT_IDX)`: 사용금액의 상대성 지표


- `건수지표(CASCNT_IDX)`: 사용건수의 상대성 지표

```python
# Columns Feature Engineering
def col_engineer(df):

    # (칼럼 생성)

    # '년월' 칼럼 생성
    df['년월'] = df['col'].str.split('^').str[0]
    # '연도' 칼럼 생성
    df['연도'] = df['년월'].str[:4]
    # '월' 칼럼 생성
    df['월'] = df['년월'].str[5:]

    # '기준월' 칼럼 생성 
    cols = ['연도', '월']
    df['기준월'] = df[cols].apply(lambda row: '-'.join(row.values.astype(str)), axis=1)

    # '업종_대분류' 칼럼 생성
    df['업종_대분류'] = df['col'].str.split('^').str[1]
    # '업종_중분류' 칼럼 생성
    df['업종_중분류'] = df['col'].str.split('^').str[2]
    # '연령대' 칼럼 생성
    df['연령대'] = df['col'].str.split('^').str[3]
    # '성별' 칼럼 생성
    df['성별'] = df['col'].str.split('^').str[4]
    # '광역시도' 칼럼 생성
    df['광역시도'] = df['col'].str.split('^').str[5]
    # '시군구' 칼럼 생성
    df['시군구'] = df['col'].str.split('^').str[6]
    # '금액지표' 칼럼 생성
    df['금액지표'] = df['col'].str.split('^').str[7]
    # '건수지표' 칼럼 생성
    df['건수지표'] = df['col'].str.split('^').str[8]

    # 불필요한 칼럼 삭제 
    df.drop(['col', '년월'],axis=1, inplace=True)

    # 칼럼 순서 변경 
    df = df[['기준월', '연도', '월', '업종_대분류', '업종_중분류', '연령대', '성별', '광역시도', '시군구', '금액지표', '건수지표']]

    return df 
```

```python
# col_engineer 함수에 데이터셋 적용 
df_2001 = col_engineer(df_2001)
df_2002 = col_engineer(df_2002)
df_2003 = col_engineer(df_2003)

df_2101 = col_engineer(df_2101)
df_2102 = col_engineer(df_2102)
df_2103 = col_engineer(df_2103)

df_2201 = col_engineer(df_2201)
df_2202 = col_engineer(df_2202)
df_2203 = col_engineer(df_2203)
```

```python
# '원본_기준일' 칼럼 값을 '연도'/'월'/'일' 칼럼으로 분리 
def date_col_engineer(df):

    # '연도' 칼럼 생성 
    df['연도'] = df['원본_기준일'].str.split('.').str[0]
    # '월' 칼럼 생성 
    df['월'] = df['원본_기준일'].str.split('.').str[1]
    # '일' 칼럼 생성 
    df['일'] = df['원본_기준일'].str.split('.').str[2]

    # '원본_기준일' 칼럼 삭제 
    df.drop(['원본_기준일'],axis=1, inplace=True)

    return df 
```

```python
# date_col_engineer 함수에 데이터셋 적용 
seoul_all = date_col_engineer(seoul_all)
seoul_gu_tidy = date_col_engineer(seoul_gu_tidy) 
```

```python
# 2020년 ~ 2022년 3월 데이터 추출 
def year_select(df):

    # '연도' 칼럼에서 '20'인 값을 '2020'으로 변경
    df['연도'] = df['연도'].str.replace('^20$', '2020') 

    # 데이터 타입 변경
    df[['연도', '월', '일']] = df[['연도', '월', '일']].astype(int)

    # '기준일' 칼럼 생성
    cols = ['연도', '월', '일']
    df['기준일'] = df[cols].apply(lambda row: '-'.join(row.values.astype(str)), axis=1)

    # 2020년 ~ 2022년 3월 데이터 추출 
    # 2022년 4월 데이터 삭제 
    filter = (df['연도'] == 2022) & (df['월'] == 4)
    index_4 = df[filter].index
    df.drop(index_4, inplace=True)

    # 인덱스 기준 내림차순 정렬
    df = df.sort_index(ascending=False)

    # 인덱스 초기화
    df.reset_index(drop=True, inplace=True)

    return df
```

```python
# 2020년 4월 21일까지의 '전국 추가 확진' 칼럼 값 계산하기 
def calcul(df):

  for i in df[df['기준일'] < '2020-04-22'].index:
    if i > 0:
      df['전국 추가 확진'][i] = df['전국 확진'][i] - df['전국 확진'][i-1]

      if df['전국 추가 확진'][i] < 0:
        df['전국 추가 확진'][i] = 0

    else:
      df['전국 추가 확진'][i] = df['전국 확진'][i] 

  return df

# calcul 함수에 crn_all 데이터 적용
crn_all = calcul(crn_all)
```

```python
# 2020년 4월 21일까지의 '서울시 추가 확진' 칼럼 값 계산하기 
def calcul_seoul(df):

  for i in df[df['기준일'] < '2020-04-22'].index:
    if i > 0:
      df['서울시 추가 확진'][i] = df['서울시 확진자'][i] - df['서울시 확진자'][i-1] 

      if df['서울시 추가 확진'][i] < 0:
        df['서울시 추가 확진'][i] = 0
        
    else:
      df['서울시 추가 확진'][i] = df['서울시 확진자'][i] 

  return df

# calcul_seoul 함수에 crn_all 데이터 적용
crn_all = calcul_seoul(crn_all) 
```
